Write a Python program that converts text to lowercase, removes punctuation and numbers, extracts all hashtags and email addresses using regex, and prints the cleaned text.



```python
import re
import string

text = "Hello there! This is Shahriyar, contact me at DataGlimpse.tech@gmail.com #Python #AI1234! Visit https://dataglimpse.tech."

# Convert to lowercase
text_lower = text.lower()

# Remove punctuation and numbers
clean_text = re.sub(f"[{string.punctuation}\d]", "", text_lower)

# Extract hashtags and emails
hashtags = re.findall(r"#\w+", text)
emails = re.findall(r"\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b", text)

print("Cleaned Text:", clean_text.strip())
print("Hashtags:", hashtags)
print("Emails:", emails)
```

Question 2
Create a Python program to extract all @mentions and URLs from a social media post using re.findall().



```python
import re

post = "Excited to collaborate with @DataGlimpse on our new project! Visit https://dataglimpse.tech or follow us @AI_Team."

# Extract @mentions and URLs
mentions = re.findall(r"@\w+", post)
urls = re.findall(r"https?://[^\s]+", post)

print("Mentions:", mentions)
print("URLs:", urls)
```

Question 3

Write a program to remove punctuation and numbers from a paragraph, split it into words, and count the frequency of each unique word.

```python
import re
import string
from collections import Counter

paragraph = "DataGlimpse is leading in AI in 2025! AI helps businesses grow faster, smarter, and better."

# Remove punctuation and numbers
clean_text = re.sub(f"[{string.punctuation}\d]", "", paragraph.lower())

# Split into words
words = clean_text.split()

# Count frequency
word_freq = Counter(words)

print("Word Frequency:")
for word, freq in word_freq.items():
    print(f"{word}: {freq}")
```

Question 4

Given a text containing multiple dates in formats like 12/10/2025 or 2025-10-12, extract all valid dates using regex.

```python
import re

text = "Important dates: 12/10/2025, 2025-10-12, and 10-12-25 (invalid)."

# Extract dates in both formats
dates = re.findall(r"\b(?:\d{2}/\d{2}/\d{4}|\d{4}-\d{2}-\d{2})\b", text)

print("Extracted Dates:", dates)
```

Question 5

Write a Python script to check whether each string in a list is a valid email address using regex.

```python
import re

emails = [
    "hello@dataglimpse.tech",
    "user.name@gmail.com",
    "invalid_email@com",
    "test@domain.org"
]

# Email validation regex
pattern = r"^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}$"

for email in emails:
    if re.match(pattern, email):
        print(f"{email} → Valid")
    else:
        print(f"{email} → Invalid")
```
